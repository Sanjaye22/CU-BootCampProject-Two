```python
 #Dependencies
import pandas as pd
```

```python
#path to CSV data
file = "../Data/2010-2015 datasets_Kaggle.csv"

# Read CSV data into pandas
df = pd.read_csv(file)
df.head()
```

```python
kgupdated_df = df[["State", "Uninsured Rate (2010)", "Uninsured Rate (2015)","Uninsured Rate Change (2010-2015)"]]
kgupdated_df["State"] = kgupdated_df["State"].str.strip()
kgupdated_df.head()
```

```python
# #path to CSV data
file1 = "../Data/2010_infant_mortality.csv"

# Read CSV data into pandas
df1 = pd.read_csv(file1)
df1.head()
```

```python
newdf2010_df = df1[['State', 'Deaths', 'Births', 'Death']]
newdf2010_df["State"] = newdf2010_df["State"].str.strip()
newdf2010_df["State"] = newdf2010_df["State"].str.replace(",", " ")
newdf2010_df
```

```python
newdf2010updated_df = newdf2010_df.rename(columns={"Deaths":"Deaths (2010)", "Births":"Births (2010)", "Death": "Death Per 1000 (2010)"})

newdf2010updated_df.head()
```

```python
#path to CSV data
file2 = "../Data/2015_infant_mortality.csv"

# Read CSV data into pandas
df2 = pd.read_csv(file2)
df2.head(30)
```

```python
newdf2015_df = df2[['State', 'Deaths', 'Births', 'Death']]
newdf2015_df["State"] = newdf2015_df["State"].str.strip()
newdf2015_df["State"] = newdf2015_df["State"].str.replace(",", " ")
newdf2015_df
```

```python
newdf2015updated_df = newdf2010_df.rename(columns={"Deaths":"Deaths (2015)", "Births":"Births (2015)", "Death": "Death Per 1000 (2015)"})
newdf2015updated_df.head()
```

```python
#path to CSV data
file3 = "../Data/States LatandLong.csv"

# Read CSV data into pandas
df3 = pd.read_csv(file3)
df3.head()
```

```python
latlong_df = df3[["Latitude", "Longitude", "State"]]
latlong_df
```

```python
#merge 2010 and 2015 infant mortality data by on State column
merged_df = pd.merge(newdf2010updated_df, newdf2015updated_df, how="outer", on="State")
merged_df
```

```python
#merge new combined infant mortality data and latlong data by on State column
pd.set_option('display.max_rows', None)
newmerged_df = pd.merge(merged_df, latlong_df, how="outer", on="State", indicator=True)
newmerged_df
```

```python
#merge onto new merged df and unissued data from Kaggle
finalnewmerged_df = pd.merge(newmerged_df, kgupdated_df, how="outer", on="State")

#drop _merge coloumn
finalnewmerged1_df= finalnewmerged_df.drop(["_merge"], axis=1) 
finalnewmerged1_df
```

```python
#Saved Final Cleaned Data
finalnewmerged1_df.to_csv(r'../Data/FinalCleanedData.csv', index = False)
```
