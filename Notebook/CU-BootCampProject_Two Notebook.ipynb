```python
 #Dependencies
import pandas as pd
```

```python
#path to CSV data
file = "../Data/2010-2015 datasets_Kaggle.csv"

# Read CSV data into pandas
df = pd.read_csv(file)
df.head()
```

```python
# Cleaning insurance data by selecting specific columns 
insurance_df = df.iloc[:, 0:4]
insurance_df.head()
```

```python
# Renamed columns in insurance dataframe
newinsurance_df = insurance_df.rename(columns={"Uninsured Rate (2010)":"Uninsured_Rate_2010", "Uninsured Rate (2015)":"Uninsured_Rate_2015", 
                                               "Uninsured Rate Change (2010-2015)": "Uninsured_Rate_Change_2010-2015"})
newinsurance_df.head()
```

```python
# Cleaning data by removing the white spaces 
newinsurance_df = insurance_df.rename(columns={"Uninsured Rate (2010)":"Uninsured_Rate_2010", "Uninsured Rate (2015)":"Uninsured_Rate_2015", 

                                               "Uninsured Rate Change (2010-2015)": "Uninsured_Rate_Change_2010-2015"})

newinsurance_df["State"] = newinsurance_df["State"].str.strip()
newinsurance_df.head()
```

```python
# Path to file to read CSV data into pandas
filestateinfant1 = "../Data/2010_infant_mortality.csv"

# Read CSV data to create the 2010 states infants dataframe
df1 = pd.read_csv(filestateinfant1)
df1.head()
```

```python
# Cleaning 2010 States infants data 
stateinfant2010_df = df1.iloc[:, 0:5]

stateinfant2010_df.head()
```

```python
# Renamed columns in 2010 state infants dataframe
newstateinfant2010_df =stateinfant2010_df.rename(columns={"Deaths":"Deaths_2010", "Births":"Births_2010", 
                                                          "State,Code":"State_ID", 
                                                              "Death": "Death_Per_1000_2010"})
newstateinfant2010_df.head()
```

```python
newstateinfant2010_df["State"] = newstateinfant2010_df["State"].str.strip()
newstateinfant2010_df["State"] = newstateinfant2010_df["State"].str.replace(",", " ")
newstateinfant2010_df.head()
```

```python
# Path to file to read CSV data into pandas
filestateinfant2 = "../Data/2015_infant_mortality.csv"

# Read CSV data to create the 2015 states infants dataframe
df2 = pd.read_csv(filestateinfant2)
df2.head()
```

```python
# Cleaning 2015 States infants data 
stateinfant2015_df = df2.iloc[:,0:5]

stateinfant2015_df.head()
```

```python
# Renamed columns in 2015 state infants dataframe
newstateinfant2015_df = stateinfant2015_df.rename(columns={"Deaths":"Deaths_2015", "Births":"Births_2015", 
                                                           "State,Code":"State_ID", "Death": "Death_Per_1000_2015"})
newstateinfant2015_df.head()
```

```python
newstateinfant2015_df["State"] = newstateinfant2015_df["State"].str.strip()
newstateinfant2015_df["State"] = newstateinfant2015_df["State"].str.replace(",", " ")
newstateinfant2015_df.head()
```

```python
# Path to file to read CSV data into pandas
statslatlong = "../Data/States LatandLong.csv"

# Read CSV data to create the states dataframe with latitudes and Longtitudes
latlong_df = pd.read_csv(statslatlong)
latlong_df.head()
```

```python
# Merge 2010 and 2015 states infant mortality data on State column
merged_df = pd.merge(newstateinfant2010_df, newstateinfant2015_df, how="outer", on="State")
merged_df.head()
```

```python
# Merge new combined infant mortality data and latlong data by on State column
pd.set_option('display.max_rows', None)
newmerged_df = pd.merge(merged_df, latlong_df, how="outer", on="State", indicator=True)
newmerged_df.head()
```

```python
# Merge states infant data and latlong data with insurance data from Kaggle
finalnewmerged_df = pd.merge(newmerged_df, newinsurance_df, how="outer", on="State")

# Drop _merge coloumn
finalnewmerged1_df= finalnewmerged_df.drop(["_merge", "State_ID_x", "State_ID_y"], axis=1) 
finalnewmerged1_df.head()
```

```python
finalnewmerged1_df.dtypes
```

```python
newfile_df = finalnewmerged1_df

newfile_df[['Death_Per_1000_2010', 'Death_Per_1000_2015']] = newfile_df[['Death_Per_1000_2010', 'Death_Per_1000_2015']].astype(int) 
newfile_df.head()
```

```python
newfile_df.dtypes
```

```python
#Saved Final Cleaned Data
newfile_df.to_csv(r'../Data/acadata.csv', index = False)
```
